```python
import pandas as pd
```

# VERI

```python
df_cleaned = pd.read_csv('cleaned_veri_data.csv')
df_cleaned['FirstActivityTime'] = pd.to_datetime(df_cleaned['FirstActivityTime'])
df_cleaned['LastActivityTime'] = pd.to_datetime(df_cleaned['LastActivityTime'])
```

```python
df_t  = df_cleaned[(df_cleaned['FirstActivityTime'] >= '2022-03-11') & (df_cleaned['FirstActivityTime'] <= '2022-03-14')]
df_t = df_t.drop(columns=['Unnamed: 0'])
```

```python
df_t[(df_t['Balance'] >= 0) & (df_t['NumTransactions']>1) & (df_t['Address']!='tz1c31KQdndxSMon9umQu3XQpLdzqTJ8FZ68') &
    (df_t['LastActivityTime'] > df_t['FirstActivityTime'])].describe()
```

```python
df_t[(df_t['Balance'] > 0) & (df_t['NumTransactions']>1) & 
     (df_t['Address']!='tz1c31KQdndxSMon9umQu3XQpLdzqTJ8FZ68')].sort_values(['NumTransactions'], ascending=False)
```
